### ノック５１：物流に関するデータを読み込んでみよう

```python
import pandas as pd

factories = pd.read_csv('tbl_factory.csv', index_col=0)
factories
```

```python
warehouses = pd.read_csv('tbl_warehouse.csv', index_col=0)
warehouses
```

```python
cost = pd.read_csv('rel_cost.csv', index_col=0)
cost.head()
```

```python
trans = pd.read_csv('tbl_transaction.csv', index_col=0)
trans.head()
```

```python
join_data = pd.merge(trans, cost, left_on=['ToFC', 'FromWH'], right_on=['FCID', 'WHID'], how='left')
join_data.head()
```

```python
join_data = pd.merge(join_data, factories, left_on='ToFC', right_on='FCID', how='left')
join_data.head()
```

```python
join_data = pd.merge(join_data, warehouses, left_on='FromWH', right_on='WHID', how='left')
join_data = join_data[['TransactionDate', 'Quantity', 'Cost', 'ToFC', 'FCName', 'FCDemand', 'FromWH', 'WHName', 'WHSupply', 'WHRegion']]
join_data.head()
```

```python
kanto = join_data.loc[join_data['WHRegion'] == '関東']
kanto.head()
```

```python
tohoku = join_data.loc[join_data['WHRegion'] == '東北']
tohoku.head()
```

### ノック５２：現状の輸送量、コストを確認してみよう

```python
print('関東支社の総コスト：' + str(kanto['Cost'].sum()) + '万円')
print('東北支社の総コスト：' + str(tohoku['Cost'].sum()) + '万円')
```

```python
print('関東支社の総部品輸送個数：{}個'.format(kanto['Quantity'].sum()))
print('東北支社の総部品輸送個数：{}個'.format(tohoku['Quantity'].sum()))
```

```python
tmp = (kanto['Cost'].sum() / kanto['Quantity'].sum()) * 10000
print('関東支社の部品1つ当たりの輸送コスト：{}円'.format(int(tmp)))
tmp = (tohoku['Cost'].sum() / tohoku['Quantity'].sum()) * 10000
print('東北支社の部品1つ当たりの輸送コスト：{}円'.format(int(tmp)))
```

```python
cost_chk = pd.merge(cost, factories, on='FCID', how='left')
print('関東支社の平均輸送コスト：{}万円'.format(cost_chk[cost_chk['FCRegion']=='関東']['Cost'].mean()))
print('東北支社の平均輸送コスト：{}万円'.format(cost_chk[cost_chk['FCRegion']=='東北']['Cost'].mean()))
```

### ノック５３：ネットワークを可視化してみよう

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

#Create Graphics object
G=nx.Graph()

# Set vertex
G.add_node('nodeA')
G.add_node('nodeB')
G.add_node('nodeC')

# Set side
G.add_edge('nodeA', 'nodeB')
G.add_edge('nodeA', 'nodeC')
G.add_edge('nodeB', 'nodeC')

# Set the point
pos ={}
pos['nodeA']=(0,0)
pos['nodeB']=(1,1)
pos['nodeC']=(0,1)

# paint
nx.draw_networkx(G, pos)

# show
plt.show()
```

### ノック５４：ネットワークにノードを追加してみよう

```python
G.add_node('nodeD')

G.add_edge('nodeA', 'nodeD')

pos['nodeD'] = (1,0)

nx.draw_networkx(G, pos, with_label=True)
```

### ノック５５：ルートの重みづけを実施しよう

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# reading data
df_w = pd.read_csv('network_weight.csv')
df_p = pd.read_csv('network_pos.csv')

# listable edge weight
size = 10
edge_weights = []
for i in range(len(df_w)):
    for j in range(len(df_w.columns)):
        edge_weights.append(df_w.iloc[i][j]*size)

# create Graphic object
G = nx.Graph()

# Setting vertex
for i in range(len(df_w.columns)):
    G.add_node(df_w.columns[i])

# Setting edge
for i in range(len(df_w.columns)):
    for j in range(len(df_w.columns)):
        G.add_edge(df_w.columns[i], df_w.columns[j])

# Setting points
pos = {}
for i in range(len(df_w.columns)):
    node = df_w.columns[i]
    pos[node] = (df_p[node][0], df_p[node][1])

# draw
nx.draw_networkx(G, pos, with_label=True, font_size=16, node_size=1000, node_color='k', font_color='w', width=edge_weights)

# show
plt.show()
```

### ノック５７：輸送ルート情報からネットワークを可視化してみよう

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

df_tr = pd.read_csv('trans_route.csv', index_col='工場')
df_pos = pd.read_csv('trans_route_pos.csv')

G = nx.Graph()

for i in range(len(df_pos.columns)):
    G.add_node(df_pos.columns[1])

num_pre = 0
edge_weights =[]
size = 0.1
for i in range(len(df_pos.columns)):
    for j in range(len(df_pos.columns)):
        if not (i==j):
            # edge
            G.add_edge(df_pos.columns[i], df_pos.columns[j])
            # edge weight
            if num_pre<len(G.edges):
                num_pre = len(G.edges)
                weight = 0
                if (df_pos.columns[i] in df_tr.columns)and(df_pos.columns[j] in df_tr.index):
                    if df_tr[df_pos.columns[i]][df_pos.columns[j]]:
                        weight = df_tr[df_pos.columns[i]][df_pos.columns[j]]*size
                elif(df_pos.columns[j] in df_tr.columns)and(df_pos.columns[i] in df_tr.index):
                    if df_tr[df_pos.columns[j]][df_pos.columns[i]]:
                        weight = df_tr[df_pos.columns[j]][df_pos.columns[i]]*size
                edge_weights.append(weight)
                
pos = {}
for i in range(len(df_pos.columns)):
    node = df_pos.columns[i]
    pos[node] = (df_pos[node][0], df_pos[node][1])
    
nx.draw(G, pos, with_labels=True, font_size=16, node_size=1000, node_color='k', font_color='w', width=edge_weights)

plt.show()
```

### ノック５８：輸送コスト関数を作成しよう

```python
import pandas as pd

df_tr = pd.read_csv('trans_route.csv', index_col='工場')
df_tc = pd.read_csv('trans_cost.csv', index_col='工場')

def trans_cost(dr_tr, df_tc):
    cost = 0
    for i in range(len(df_tc.index)):
        for j in range(len(df_tr.columns)):
            cost += df_tr.iloc[i][j]*df_tc.iloc[i][j]
    return cost
    
print('総輸送コスト：{}'.format(trans_cost(df_tr, df_tc)))
```

### ノック５９：制約条件を作ってみよう

```python
import pandas as pd

df_tr = pd.read_csv('trans_route.csv', index_col='工場')
df_demand = pd.read_csv('demand.csv')
df_supply = pd.read_csv('supply.csv')

for i in range(len(df_demand.columns)):
    temp_sum = sum(df_tr[df_demand.columns[i]])
    print('{}への輸送量{}（需要量：{}）'.format(df_demand.columns[i], temp_sum, df_demand.iloc[0][i]))
    if temp_sum>=df_demand.iloc[0][i]:
        print('需要量を満たしています。')
    else:
        print('需要量を満たしていません。輸送ルートを再計算してください。')
        
for i in range(len(df_supply.columns)):
    temp_sum = sum(df_tr.loc[df_supply.columns[i]])
    print('{}からの輸送量：{}（供給限界：{}）'.format(df_supply.columns[i], temp_sum, df_supply.iloc[0][i]))
    if temp_sum<=df_supply.iloc[0][i]:
        print('供給限界の範囲内です。')
    else:
        print('供給限界を超過しています。輸送ルートを再計算してください。')
```

### ノック６０：輸送ルートを変更して、輸送コスト関数の変化を確認しよう

```python
import pandas as pd
import numpy as np

df_tr_new = pd.read_csv('trans_route_new.csv', index_col='工場')
print(df_tr_new)

print('総輸送コスト(変更後):'+str(trans_cost(df_tr_new, df_tc)))

def condition_demand(df_tr, df_demand):
    flag = np.zeros(len(df_demand.columns))
    for i in range(len(df_demand.columns)):
        temp_sum = sum(df_tr[df_demand.columns[i]])
        if (temp_sum>=df_demand.iloc[0][i]):
            flag[i] = 1
    return flag

def condition_supply(df_tr, df_supply):
    flag = np.zeros(len(df_supply.columns))
    for i in range(len(df_supply.columns)):
        temp_sum = sum(df_tr.loc[df_supply.columns[i]])
        if temp_sum<=df_supply.iloc[0][i]:
            flag[i] = 1
    return flag

print('需要条件計算結果:'+str(condition_demand(df_tr_new, df_demand)))
print('供給条件計算結果:'+str(condition_supply(df_tr_new, df_supply)))
```
